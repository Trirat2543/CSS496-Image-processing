```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/skz.jpg',0)
```

```python
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img2 = cv2.imread('/content/skz.jpg')
cv2_imshow(img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_gray = cv2.imread('/content/skz.jpg',0)
img_color = cv2.applyColorMap(img_gray, cv2.COLORMAP_COOL)
cv2_imshow(img_gray)
cv2_imshow(img_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
